```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data
CVs = [1, 2, 4, 8]
methods = ['FlashScheduler', 'DefaultScheduler']
colors=['#df3f31','#143371','#7ea34a']  # Specify colors for methods
modelname="bert"
# Initialize lists to store the average TimeDifference values
average_time_difference_csv1 = []
average_time_difference_csv2 = []

# Load and process data for each CV
for cv in CVs:
    # Load the CSV files for both methods/home/flash/evaluation/metrics/cvd/bert-fwm/DefaultScheduler/1695135009
    # /home/flash/evaluation/metrics/cvd/bert-fwm/FlashScheduler/1695051705  former
    csv1 = pd.read_csv(f'/home/flash/evaluation/metrics/cvd/{modelname}-fwm/FlashScheduler/1695245496/{cv}/{modelname}-fwm.csv')
    csv2 = pd.read_csv(f'/home/flash/evaluation/metrics/cvd/{modelname}-fwm/DefaultScheduler/1695104023/{cv}/{modelname}-fwm.csv')

    # Filter rows where 'StatusCode' is 200
    csv1 = csv1[csv1['StatusCode'] == 200]
    csv2 = csv2[csv2['StatusCode'] == 200]

    # Calculate TimeDifference in seconds
    csv1['TimeDifference'] = (csv1['ResponseTime'] - csv1['RequestTime']) / 1e9  # Convert nanoseconds to seconds
    csv2['TimeDifference'] = (csv2['ResponseTime'] - csv2['RequestTime']) / 1e9  # Convert nanoseconds to seconds

    # Calculate and append the average TimeDifference values
    average_time_difference_csv1.append(csv1['TimeDifference'].mean())
    average_time_difference_csv2.append(csv2['TimeDifference'].mean())

# Create a bar chart to compare the methods with consistent colors
width = 0.4
x = np.arange(len(CVs))
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each method's data separately
for i, method in enumerate(methods):
    ax.bar(x + i * width, [avg for avg in (average_time_difference_csv1, average_time_difference_csv2)[i]],
           width=width, label=method, color=colors[i])

ax.set_xlabel('CV Values')
ax.set_ylabel('Average latency (seconds)')
ax.set_title(f'Average latency for Different CV Values of {modelname}')
ax.set_xticks(x + width / 2)
ax.set_xticklabels(CVs)
ax.legend()

plt.show()
```
