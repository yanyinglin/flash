```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Data
CVs = [1, 2,4,8]
methods = ['FlashScheduler', 'DefaultScheduler', 'INFlessScheduler']
labels = ['Flash', 'no cpu no sorted', 'Infless']
colors = ['#df3f31', '#143371', '#7ea34a']  # Specify colors for methods
model_timestamps = {
    "labse":[1695234665,1695144132,1695290860],
    "bert-qa": [1695238221, 1695139547,1695294389],
    "vggnet-11": [1695241700,1695130532,1695297955],
    "bert":[1695245496,1695104023,1695301807],
    "resnet-152":[1695249276,1695121357,1695305355],
    "resnet-101":[1695252824,1695116824,1695308835],
    "resnet-50":[1695256342,1695111942,1695312819],
    "shufflenet":[1695259836,1695271992,1695316225]
}

# Create a dictionary to store data for all models
all_model_data = {}

for modelname, timestamps in model_timestamps.items():
    model_data = {}
    
    for i, method in enumerate(methods):
        timestamp = timestamps[i]
        method_CVS_response = []
        
        for cv in CVs:
            csv = pd.read_csv(f'/home/flash/evaluation/metrics/cvd/{modelname}-fwm/{method}/{timestamp}/{cv}/{modelname}-fwm.csv')
            
            # Modify this line:
            csv_filter = csv[csv['StatusCode'] == 200].copy()
            csv_filter['TimeDifference'] = (csv_filter['ResponseTime'] - csv_filter['RequestTime']) / 1e9
            method_CVS_response.append(csv_filter['TimeDifference'].values)
        
        model_data[method] = method_CVS_response
    print(model_data)
    # Store the data for the current model in the all_model_data dictionary
    all_model_data[modelname] = model_data

    # Saving the data for the current model to a CSV file
    save_directory = f'/home/flash/evaluation/plots/latency/{modelname}'
    os.makedirs(save_directory, exist_ok=True)
    for method, method_CVS_response in model_data.items():
        # Flatten the list of lists into a single list
        flattened_data = [item for sublist in method_CVS_response for item in sublist]
        
        # Create a DataFrame with a single column 'TimeDifference'
        method_df = pd.DataFrame({'TimeDifference': flattened_data})
        
        # Specify the filename
        filename = f'{modelname}_{method}_combined.csv'
        
        # Combine the save_directory and filename to get the save path
        save_path = f'{save_directory}/{filename}'
        
        # Save the method's data to a CSV file
        method_df.to_csv(save_path, index=False)


# Now you have all the data for all models stored in all_model_data dictionary
```
