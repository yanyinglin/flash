```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

modelname = 'bert'
methods = ['FlashScheduler', 'DefaultScheduler', 'INFlessScheduler']
labels=['Flash','Default','INFless']
colors=['#df3f31','#143371','#01d06c']
# Define the directory where the CSV files are located
directory = f'/home/flash/evaluation/plots/{modelname}/'

# Create a list to store data from the CSV files
data_list = []

# Loop through the methods and read data from each CSV file
for method in methods:
    csv_file = os.path.join(directory, f'{modelname}_{method}_combined.csv')
    df = pd.read_csv(csv_file)
    data_list.append(df['TimeDifference'])

# Create an ECDF plot for each dataset
plt.figure(figsize=(5, 2.6))
# Set the style to "whitegrid" to enable grid lines
sns.set(style="whitegrid")

for i, data in enumerate(data_list):
    sns.ecdfplot(data=data, label=labels[i],color=colors[i])

# Add labels and legend
plt.xlabel(f'{modelname} Response time',fontsize=12)
plt.ylabel('CDF',fontsize=14)
plt.legend(title='Methods',fontsize=13)
plt.xticks(fontsize=12)  
plt.yticks(fontsize=12)  
plt.show()
```
