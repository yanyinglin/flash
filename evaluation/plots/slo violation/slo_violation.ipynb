```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

CVs = [1, 2, 4, 8]
methods = ['FlashScheduler', 'DefaultScheduler']

colors = ['red', 'black']  # Specify colors for methods

model_timestamps = {
    "labse": [1695202742, 1695144132],
    "bert-qa": [1695238221, 1695139547],
    "vggnet-11": [1695241700, 1695130532],
    "bert": [1695245496, 1695279460],
    "resnet-152": [1695249276, 1695121357],
    "resnet-101": [1695252824, 1695116824],
    "resnet-50": [1695256342, 1695111942],
    "shufflenet": [1695259836, 1695271992],
}

# Initialize dictionaries to store the total StatusCode 500 counts for each model and method
total_counts = {model: {method: 0 for method in methods} for model in model_timestamps}

# Loop through model_timestamps dictionary
for modelname, timestamps in model_timestamps.items():
    # Load and process data for each CV
    for cv in CVs:
        for i, method in enumerate(methods):
            timestamp = timestamps[i]
            csv = pd.read_csv(
                f'/home/flash/evaluation/metrics/cvd/{modelname}-fwm/{method}/{timestamp}/{cv}/{modelname}-fwm.csv')

            # Filter rows where 'StatusCode' is 500 and get the count
            count = len(csv[csv['StatusCode'] == 500])

            # Add the count to the total for the corresponding model and method
            total_counts[modelname][method] += count

# Extract the total counts for each method
flash_scheduler_counts = [total_counts[model][methods[0]] for model in model_timestamps]
default_scheduler_counts = [total_counts[model][methods[1]] for model in model_timestamps]

# Create a bar chart to compare the total StatusCode 500 counts for each model and method
x = np.arange(len(model_timestamps))
width = 0.4
fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(x - width/2, flash_scheduler_counts, width=width, label='FlashScheduler', color='red')
ax.bar(x + width/2, default_scheduler_counts, width=width, label='DefaultScheduler', color='black')

ax.set_xlabel('Models')
ax.set_ylabel('Total StatusCode 500 Count')
ax.set_title('Total StatusCode 500 Count for Different Models and Methods')
ax.set_xticks(x)
ax.set_xticklabels(model_timestamps.keys())
ax.legend()

plt.show()
```
