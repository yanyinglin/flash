```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# CSV
data = pd.read_csv('/home/flash/motivation/combined_function_data.csv')

# seaborn
plt.figure(figsize=(10
                    , 4))

# # 
# sns.scatterplot(x='cpu_cores', y='exec_time', hue='function_name', data=data, marker='o', s=50)

# for model in data:
#     model=data['function_name']
# print(model)
# model_name = model.replace("-fwm", "")
    # print(model_name)
data['function_name_fwm']=data['function_name'].str.rstrip('-fwm')
print(data['function_name_fwm'])
sns.pointplot(x='cpu_cores', y='exec_time', hue='function_name_fwm', data=data,              
              markers="o", scale=0.8)
plt.xlabel('CPU Cores',fontsize=14)
plt.ylabel('Execution Time (S)',fontsize=14)

plt.title('Execution Time vs CPU Cores',fontsize=16)
plt.legend(fontsize=14,ncol=2,handlelength=0, handletextpad=0.35)
plt.xticks(fontsize=12)  
plt.yticks(fontsize=12) 
plt.grid(True)
plt.savefig('interference_cdf.pdf',bbox_inches='tight',dpi=120)
plt.show()
```
